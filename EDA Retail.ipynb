```python
# Visualize profit by category
plt.figure(figsize=(12, 6))
sns.barplot(x='Category', y='Profit', data=df)
plt.title('Profit by Category')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualize profit by sub-category
plt.figure(figsize=(16, 8))
sns.barplot(x='Sub-Category', y='Profit', data=df)
plt.title('Profit by Sub-Category')
plt.xticks(rotation=90)
plt.show()
```

```python
# Visualize profit by ship mode
plt.figure(figsize=(10, 5))
sns.barplot(x='Ship Mode', y='Profit', data=df)
plt.title('Profit by Ship Mode')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualize profit by segment
plt.figure(figsize=(10, 5))
sns.barplot(x='Segment', y='Profit', data=df)
plt.title('Profit by Segment')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualize profit by state
plt.figure(figsize=(16, 8))
sns.barplot(x='State', y='Profit', data=df)
plt.title('Profit by State')
plt.xticks(rotation=90)
plt.show()
```

```python
# Example: Histogram of Profit
plt.figure(figsize=(10, 6))
sns.histplot(df['Profit'], bins=30, kde=True)
plt.title('Distribution of Profit')
plt.xlabel('Profit')
plt.ylabel('Frequency')
plt.show()
```

```python
# Example: Bar plot of Sales by Region
plt.figure(figsize=(12, 6))
sns.barplot(x='Region', y='Sales', data=df)
plt.title('Sales by Region')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
# Pairplot for numerical variables
import numpy as np
numerical_vars = df.select_dtypes(include=np.number)
sns.pairplot(numerical_vars)
plt.show()
```

```python
# Example: Correlation heatmap
correlation_matrix = df.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
